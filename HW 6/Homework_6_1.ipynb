### Step 0

Run the necessary imports.

```python
import dlib
import cv2
import matplotlib.pyplot as plt
import urllib.request
```

### Step 1

Load an image (any image that contains faces).

```python
url = 'https://cdn.marvel.com/content/2x/001avn_ons_mas_mob_01_0.jpg'
urllib.request.urlretrieve(url, "avengers.jpg")

img = cv2.imread('avengers.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img)
```

### Step 2

Load the dlib face predictor.

```python
import dlib
detector = dlib.get_frontal_face_detector()
```

### Step 3

Run the predictor on your image.

```python
rects = detector(gray, 1)

print('Number of detected faces:', len(rects))
print(rects)
print(rects[0].left)
```

### Step 4

Draw bounding boxes around the detected faces and plot the image. Use different colour for each face.

```python
import numpy as np

def rect_to_bb(rect):
    x = rect.left()
    y = rect.top()
    w = rect.right() - x
    h = rect.bottom() - y

    return (x, y, w, h)

result_dlib = np.copy(img)
faces_dlib_img = []
for rect in rects:
    x, y, w, h = rect_to_bb(rect)
    print(x, y, w, h)
    color = [np.random.randint(0, 255) for _ in range(3)]

    cv2.rectangle(result_dlib, (x, y), (x+w, y+h), color, 5)
    faces_dlib_img.append(img[y:y+h, x:x+w, :])


plt.imshow(result_dlib), plt.title('dlib')
```

### Step 5 (optional)

Repeat the process with a different and more challenging image (more faces, smaller faces, people with glasses, hats, helmets, etc.). How does the detector perform? Is it robust?

```python
from collections import namedtuple

ImageData = namedtuple('ImageData', ['url', 'filename'])

def get_img(img_data: ImageData):
    url, filename = img_data.url, img_data.filename
    full_filename = f'{filename}.jpg'

    urllib.request.urlretrieve(url, full_filename)

    img = cv2.imread(full_filename)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    return img
```

```python
def show_faces(detector, img):
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    rects = detector(gray, 1)

    result_dlib = np.copy(img)
    faces_dlib_img = []
    for rect in rects:
        x, y, w, h = rect_to_bb(rect)
        print(x, y, w, h)
        color = [np.random.randint(0, 255) for _ in range(3)]

        cv2.rectangle(result_dlib, (x, y), (x+w, y+h), color, 5)
        faces_dlib_img.append(img[y:y+h, x:x+w, :])


    plt.imshow(result_dlib), plt.title('dlib')
```
